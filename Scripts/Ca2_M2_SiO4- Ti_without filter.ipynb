```python
from pymatgen.core import Structure
import os
```

```python
os.chdir('/home/smart/Downloads')
structure = Structure.from_file('EntryWithCollCode20340.cif')
```

```python
#print (structure)
structure.replace_species({('Zr4+'):('Ti4+')})
s1 = structure.copy()
```

```python
#print(s1)
```

```python
from pymatgen.core.structure import Element
from pymatgen.transformations.standard_transformations import SubstitutionTransformation
trans = SubstitutionTransformation({'Na+':{'Ca2+':0.5}})
s2 = trans.apply_transformation(s1)
s3 = s2.copy()
#print(s3)
```

```python
from pymatgen.transformations.advanced_transformations import EnumerateStructureTransformation
d=EnumerateStructureTransformation(refine_structure=False)
e=d.apply_transformation(s3,1000)
```

```python
print(e)
```

```python
from pymatgen.analysis.structure_matcher import StructureMatcher
from pymatgen.io.vasp.sets import MPRelaxSet
from pymatgen.io.vasp.inputs import Kpoints
for l in range(0,16):
    k=Kpoints.automatic_gamma_density(e[l]["structure"],1000)
    relax_set = MPRelaxSet(e[l]["structure"], user_incar_settings={"EDIFF": 1e-05, "ENCUT": 520, "LDAU": True, "NELM": 300, "NSW": 99, "ISYM": 0, "LPLANE": True, "NCORE": 6, "NSIM": 4, "ISMEAR": 0}, user_kpoints_settings=k)
    relax_set.write_input("%d" % l)
```
