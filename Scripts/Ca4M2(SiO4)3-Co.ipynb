```python
from pymatgen.core import Structure
```

```python
import os
```

```python
os.chdir('/home/smart/Downloads')
```

```python
structure = Structure.from_file('EntryWithCollCode20340.cif')
```

```python
s1 = structure.copy()

print(s1)
```

```python
s1.replace_species({('Na+'):('Ca2+')})
s2 = s1.copy()
print(s2)
```

```python
s2.replace_species({('Zr4+'):('Co2+')})
s3= s2.copy()
```

```python
print(s3)
```

```python
from pymatgen.analysis.structure_prediction.volume_predictor import RLSVolumePredictor
p = RLSVolumePredictor(radii_type="atomic")
sorted_structure= p.get_predicted_structure(s3,structure)

from pymatgen.io.vasp.inputs import Poscar
import os
os.chdir('/home/smart/Downloads/')
sorted_structure.to('poscar', filename='POSCAR_Ca4Co2(SiO4)3')
```
